```python
import csv

yelp = []

i = 0
with open('/a.csv', 'r',  encoding='utf8') as file:
    csvreader = csv.reader(file)
    next(csvreader, None) #omitim capçalera
    
    for row in csvreader:
        i = i + 1
        yelp.append(row)
        if i == 15:
            break

```

```python
yelp
```

```python


# Crear un diccionario para almacenar los datos transformados
datos_transformados = {}

# Iterar a través de la lista bidimensional
for reseña in yelp:
    id_restaurante = reseña[0]
    fecha_inicial = reseña[1]
    fecha_final = reseña[2]
    nombre_restaurante = reseña[3]
    descripcion = reseña[4]
    direccion = reseña[5]
    calificacion = reseña[6]
    num_reseñas = reseña[7]
    tipos_cocina = reseña[8].split(', ')  # Convertir los tipos de cocina en una lista
    precio = reseña[9].replace('$', '€')  # Reemplazar el símbolo de dólar por euro en el precio
    
    # Separar el código postal del resto de la dirección
    partes_direccion = direccion.split(' ')
    zipcode = partes_direccion[-1]
    direccion = ' '.join(partes_direccion[:-1])
    
    # Crear un diccionario con la información del restaurante
    restaurante = {
        'Fecha Inicial': fecha_inicial,
        'Fecha Final': fecha_final,
        'Nombre del Restaurante': nombre_restaurante,
        'Descripción': descripcion,
        'Dirección': direccion,
        'Zipcode': zipcode,
        'Calificación': calificacion,
        'Número de Reseñas': num_reseñas,
        'Tipos de Cocina': tipos_cocina,
        'Precio': precio
    }
    
    # Verificar si ya existe una entrada para el restaurante y agregarla a la lista de comentarios si es el caso
    if id_restaurante in datos_transformados:
        datos_transformados[id_restaurante]['Comentarios'].append(descripcion)
    else:
        datos_transformados[id_restaurante] = {
            'Restaurante': restaurante,
            'Comentarios': [descripcion]
        }

# Imprimir el diccionario final
for rank, datos in datos_transformados.items():
    print(f'Rank: {rank}')
    print(f'Restaurante: {datos["Restaurante"]}')
    print(f'Comentarios: {datos["Comentarios"]}')
    print()
```
