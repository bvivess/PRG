```python
import csv

yelp = []

i = 0
with open('/a.csv', 'r',  encoding='utf8') as file:
    csvreader = csv.reader(file)
    next(csvreader, None) #omitim capçalera
    
    for row in csvreader:
        i = i + 1
        yelp.append(row)
        if i == 10:
            break
        
```
